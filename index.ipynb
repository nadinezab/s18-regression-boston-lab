### Let's do Regression 

Now, let's perform a number of simple regression experiments between the chosen independent variables and the dependent variable (price). You'll do this in a loop and in every iteration, you should pick one of the independent variables. Perform the following steps:

* Run a simple OLS regression between independent and dependent variables
* Plot a regression line on the scatter plots
* Plot the residuals using `sm.graphics.plot_regress_exog()`
* Plot a Q-Q plot for regression residuals normality test 
* Store following values in array for each iteration:
    * Independent Variable
    * r_squared'
    * intercept'
    * 'slope'
    * 'p-value'
    * 'normality (JB)' 
* Comment on each output 

```python
results = []

for var in ['crim', 'dis', 'rm', 'zn', 'age']:
    
    # define model
    
    f = f'medv ~ {var}'
    model = ols(formula=f, data=dataset).fit()
    
    # plot regression line and scatter plot
    X_new = pd.DataFrame({var: [dataset[var].min(), dataset[var].max()]});
    preds = model.predict(X_new)
    plt.scatter(dataset[var], dataset['medv'], label = 'Data Points')
    plt.plot(X_new, preds, label = 'Regression Line', color = 'blue')
    plt.title(f'Regression line for {var}')
    plt.show()
    
    # plot residuals
    fig = plt.figure(figsize=(15,8))
    fig = sm.graphics.plot_regress_exog(model, var, fig=fig)
    plt.show()
    
    # plot Q-Q plot
    residuals = model.resid
    fig = sm.graphics.qqplot(residuals, dist=stats.norm, line='45', fit=True)
    plt.title(f'Q-Q plot for {var} residuals')
    plt.show()
    
    results.append([var, model.rsquared, model.params[0], 
                    model.params[1], model.pvalues[1], sms.jarque_bera(residuals)[0]] )


df = pd.DataFrame(results)
df.columns = ['var', 'r_squared', 'intercept', 'slope', 'p-values', 'J-B normality']
df
```

```python
#Your observations here 
# Weak R-squared, best R-squared is for rm, so room number is our best predictor so far.
```

Clearly, the results are not very reliable. The best R-Squared is witnessed with `rm`, so in this analysis, this is our best predictor.
